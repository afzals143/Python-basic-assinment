### Q5
Given:
```python
A = {2, 4, 6, 8}
B = {4, 8, 12}
```
Find Union, Intersection, A-B

````python
# write your code here
A = {2, 4, 6, 8}
B = {4, 8, 12}
print("Union: ",A.union(B))
print("Intersection: ",A.intersection(B))
print("Diffrence: ",A.difference(B))
````
