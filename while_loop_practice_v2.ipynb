Q1. Print numbers from 1 to 20 using while loop.

Approach:
- Initialize a variable
- Use while condition
- Increment inside loop

```python
#write answer here
num = 1
while num<=20:
  print(num)
  num += 1
```

Q4. Find sum of numbers from 1 to n.

Approach:
- Take input n
- Add numbers in loop

```python
#write answer here
n = int(input("Enter the number: "))
sum = 0
num = 1
while num <= n:
  sum = sum + num
  num +=1

print(f"Total sum of number from 1 to {n} is: {sum}")
```

Q6. Count digits in a number.

Approach:
- Divide number by 10 repeatedly
- Count iterations

```python
#write answer here
num = int(input("Enter a number: "))

if num == 0:
    count = 1
else:

    temp_num = abs(num)
    count = 0
    while temp_num > 0:
        temp_num //= 10
        count += 1


print(f"Total digits: {count}")
```

Q7. Check if a number is palindrome.

Approach:
- Reverse number
- Compare with original

```python
#write answer here

var= int(input())
new = var
res=0
while var>0:
  temp = var % 10
  res = res*10 + temp
  var = var // 10
print(res)
if new==res:
  print("Palindrome")
else:
  print("Not Palindrome")
```
